```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
x=6*np.random.rand(200,1)-3
y=0.8*x**2+0.9*x+2+np.random.randn(200,1)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y,"b.")
plt.show()
```

```python
#Splitting the data.
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=2)
print(X_train[0])
```

```python
# Predicting the values using Linear Regression.
lr=LinearRegression()
lr.fit(X_train,Y_train)
Y_predicted=lr.predict(X_test)
r2_score(Y_test,Y_predicted)
```

```python
# Showing the Regression line graphically when applied the Linear Regression.
plt.plot(X_train,lr.predict(X_train),color="red")
plt.plot(x,y,"b.")
plt.show()
```

```python
#Applying the Polynomial Regression.
poly=PolynomialFeatures(degree=2)
X_train_transform=poly.fit_transform(X_train)
X_test_transform=poly.transform(X_test)
print(X_train[0])
print(X_train_transform[0])
```

```python
obj2=LinearRegression()
obj2.fit(X_train_transform,Y_train)
Y_predicted=obj2.predict(X_test_transform)
```

```python
r2_score(Y_test,Y_predicted)
```

```python
# Plotting the Polynomial regression graph.
x_new=(np.linspace(-3,3,200).reshape(200,1))
x_new_poly=poly.transform(x_new)
y_new=obj2.predict(x_new_poly)
```

```python
plt.plot(x_new,y_new,"r-",label="Predictions")
plt.plot(X_test,Y_test,"b.",label="Test Values")
plt.plot(X_train,Y_train,"g.",label="Training Values")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
