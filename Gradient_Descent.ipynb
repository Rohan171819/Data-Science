```python
import numpy as np
import matplotlib.pyplot as plt
def Gradient_Descent(x,y):
  m_current=b_current=0
  iterations=35
  learning_rate=0.6
  n=len(x)  # Assuming that the length og x and y are equal.
  for i in range(iterations):
    y_predicted=(m_current * x + b_current)
    cost=(1/n)*sum([values **2 for values in (y-y_predicted)])
    m_partial_derivate=-(2/n) *sum(x*(y-y_predicted))
    b_partial_derivate=-(2/n)*sum((y-y_predicted))
    m_current=m_current - learning_rate*m_partial_derivate
    b_current=b_current - learning_rate*b_partial_derivate
    print("m  {} , b  {} , cost  {} , iterations  {}".format(m_current,b_current,cost,iterations))   
x=np.array([92,56,88,70,80,49,65,35,66,67])
y=np.array([98,68,81,80,83,52,66,30,68,73])
Gradient_Descent(x,y)
fig=plt.figure()
ax=plt.axes(projection="3d")
#ax.plot3D(x,y)
ax.scatter3D(x,y)
plt.show()

```
